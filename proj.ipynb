```python
df['Income'] = df['Income'].astype(str)
df['Income'] = df['Income'].str.replace(',', '', regex = False)
df['Income'] = pd.to_numeric(df['Income'], errors = 'coerce')
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], errors='coerce')
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({'Alone':'Single', 'Absurd': 'Single', 'YOLO': 'Single'})
```

```python
df['Income'] = df['Income'].fillna(df.groupby(['Education', 'Marital_Status'])['Income'].transform('mean'))
```

```python
df['Age'] = 2025-df['Year_Birth']
df['Total_Children'] = df['Kidhome'] + df['Teenhome']
df['Amount_Spent'] = df['MntWines']+df['MntFruits']+df['MntMeatProducts']+df['MntFishProducts']+df['MntSweetProducts']+df['MntGoldProds']
purchase_cols = [
    'NumWebPurchases',
    'NumCatalogPurchases',
    'NumStorePurchases'
]
df['Total_Purchases'] = df[purchase_cols].sum(axis=1)
```

```python
plt.boxplot(df[['Income', 'Age', 'Amount_Spent', 'Total_Purchases']])
```
