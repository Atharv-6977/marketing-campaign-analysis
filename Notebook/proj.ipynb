```python
df['Income'] = df['Income'].astype(str)
df['Income'] = df['Income'].str.replace(',', '', regex = False)
df['Income'] = pd.to_numeric(df['Income'], errors = 'coerce')
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], errors='coerce')
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({'Alone':'Single', 'Absurd': 'Single', 'YOLO': 'Single'})
```

```python
df['Income'] = df['Income'].fillna(df.groupby(['Education', 'Marital_Status'])['Income'].transform('mean'))
# Whenever you want to fill missing values using group statistics, use transform(), not mean().                                                                                                                                                                                                                                                                                                                                                                                                 
```

```python
df['Age'] = 2025-df['Year_Birth']
df['Total_Children'] = df['Kidhome'] + df['Teenhome']
df['Amount_Spent'] = df['MntWines']+df['MntFruits']+df['MntMeatProducts']+df['MntFishProducts']+df['MntSweetProducts']+df['MntGoldProds']
purchase_cols = [
    'NumWebPurchases',
    'NumCatalogPurchases',
    'NumStorePurchases'
]
df['Total_Purchases'] = df[purchase_cols].sum(axis=1)
```

Older people are not as tech-savvy and probably prefer shopping in-store

Ho(Null) = Age has no relation with store purchases

H1(Alternate) = Age is positively associated with store purchases

```python
from scipy.stats import pearsonr

corr_store, p_store = pearsonr(df['Age'], df['NumStorePurchases'])
corr_web, p_web = pearsonr(df['Age'], df['NumWebPurchases'])

print(corr_store, p_store)
print(corr_web, p_web)
```

Conclusion : Older customers show a statistically significant increase in both in-store and web purchases, indicating higher overall purchasing activity rather than digital avoidance.

Customers with kids probably have less time to visit a store and would prefer to shop online

Ho(Null) =  Having kids has no relation with web purchases

H1(Alternate) = Having kids has positive relation with web purchases

```python
from scipy.stats import pearsonr

Corr_web_kids, P_web_kids = pearsonr(df['Total_Children'], df['NumWebPurchases'])
Corr_store_kids, P_store_kids = pearsonr(df['Total_Children'], df['NumStorePurchases'])
print(Corr_web_kids, P_web_kids)
print(Corr_store_kids, P_store_kids)

```

Conclusion : Reject Ho, but H1 is also not supported
Customers with kids don't have time to visit the store but this does not translate into higher online shopping activity.

Does the US fare significantly better than the rest of the world in terms of total purchases?

Ho(Null) : Mean total purchases in the US = Mean total purchases in the rest of the world

H1(Alternate) : Mean total purchases in the US ≠ Mean total purchases in the rest of the world

```python
from scipy.stats import ttest_ind

us_purchases = df[df['Country'] == 'US']['Total_Purchases']
non_us_purchases = df[df['Country'] != 'US']['Total_Purchases']

t_stat, p_val = ttest_ind(us_purchases, non_us_purchases, equal_var=False)

print(t_stat, p_val)
print(us_purchases.mean(), non_us_purchases.mean())
```

While US customers show slightly higher average purchase counts, the difference is not strong enough to conclude that the US market performs significantly better than other regions.
